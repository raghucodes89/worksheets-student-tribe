```python
#q1 create a class called student

class Student:
    def __init__(self,name,roll_number,marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks
    
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_number}")
        print(f"Marks: {self.marks}")
# create an object of the student class
student1 = Student("anil",101, 89)
# display the information of the student
student1.display_info()

```

```python
#q2 create a class called employee

class Employee:
    def __init__(self, employee_id, employee_name, salary):
        self.employee_id = employee_id
        self.name = employee_name
        self.salary = salary

    def display_details(self):
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Salary: {self.salary}")

# create an object of the employee class        
employee1 = Employee(101, "anil", 50000)
employee2 = Employee(102, "ram", 80000)
employee3 = Employee(103, "raghu", 60000)
# display the information of the employee
employee1.display_details()
employee2.display_details()
employee3.display_details()
```

```python
#q3 create a parent class called Vehicle with inheritance
class Vehicle:
    def __init__(self, name):
        self.name = name
    
    def start(self):
        print("Vehicle Started")
    
    def stop(self):
        print("Vehicle Stopped")


# Child class Car inherits from Vehicle

class Car(Vehicle):
    def __init__(self, name, model):
        super().__init__(name)
        self.model = model
    
    def drive(self):
        print("Car Driving")


# Create an object and call methods
car = Car("tata", "punch")

# Inherited method from Vehicle                 
car.start()

# Own method of Car
car.drive()
      
# Inherited method from Vehicle  
car.stop()     
```
